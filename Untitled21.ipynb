```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, confusion_matrix, roc_curve, auc
)

```

```python
df= pd.read_excel(r"Book4.xlsx")


```

```python
print("Dataset head:")
print(df.head())
print("\nColumns:", df.columns.tolist())
```

```python
plt.figure()
df['Class'].value_counts().plot(kind='bar')
plt.title("Class Distribution (0 = Normal, 1 = Fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
X = df.drop("Class", axis=1)   # Features
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
model = LogisticRegression(max_iter=200)
model.fit(X_train_scaled, y_train)
```

```python
pred = model.predict(X_test_scaled)
acc = accuracy_score(y_test, pred)

```

```python
print("\nAccuracy:", acc)
```

```python
def check_fraud(transaction_row):
    """
    Pass a single transaction (as list or DataFrame row)
    """
    row_scaled = scaler.transform([transaction_row])
    pred = model.predict(row_scaled)[0]
    return "Fraud" if pred == 1 else "Normal"
```

```python
cm = confusion_matrix(y_test, pred, labels=[0, 1])
print("\nConfusion Matrix:\n", cm)
```

```python
plt.figure()
plt.imshow(cm, interpolation='nearest')
plt.title("Confusion Matrix")
plt.colorbar()
```

```python
classes = ["Normal(0)", "Fraud(1)"]
tick_marks = np.arange(2)
plt.xticks(tick_marks, classes)
plt.yticks(tick_marks, classes)
```

```python
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(
            j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black"
        )
```

```python
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.tight_layout()
plt.show()
```

```python
y_prob = model.predict_proba(X_test_scaled)[:, 1]
fpr, tpr, threshold = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.3f}")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.title("ROC Curve – Credit Card Fraud Detection")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.tight_layout()
plt.show()
```
